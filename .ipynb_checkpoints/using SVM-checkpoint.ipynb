## Confusion Matrix

```python
conf_matrix = confusion_matrix(test_labels, test_predictions)
print("\nConfusion Matrix on Test Data:")
print(conf_matrix)

# Plot Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=list(test_generator.class_indices.keys()), 
            yticklabels=list(test_generator.class_indices.keys()))
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

```

```python
# Calculate Sensitivity (Recall), Specificity per class
def calculate_sensitivity_specificity(conf_matrix, num_classes):
    sensitivity = []
    specificity = []
    for i in range(num_classes):
        TP = conf_matrix[i, i]
        FN = conf_matrix[i, :].sum() - TP
        FP = conf_matrix[:, i].sum() - TP
        TN = conf_matrix.sum() - (TP + FN + FP)
        
        # Sensitivity (Recall)
        if (TP + FN) > 0:
            sens = TP / (TP + FN)
        else:
            sens = 0.0
        sensitivity.append(sens)
        
        # Specificity
        if (TN + FP) > 0:
            spec = TN / (TN + FP)
        else:
            spec = 0.0
        specificity.append(spec)
    
    return sensitivity, specificity

num_classes = len(test_generator.class_indices)
sensitivity, specificity = calculate_sensitivity_specificity(conf_matrix, num_classes)

# Print Sensitivity and Specificity per class
print("\nSensitivity (Recall) per class:")
for cls, sens in zip(test_generator.class_indices.keys(), sensitivity):
    print(f"{cls}: {sens:.4f}")
print("\nSpecificity per class:")
for cls, spec in zip(test_generator.class_indices.keys(), specificity):
    print(f"{cls}: {spec:.4f}")

# Average Sensitivity and Specificity
avg_sensitivity = np.mean(sensitivity)
avg_specificity = np.mean(specificity)
print(f"\nAverage Sensitivity: {avg_sensitivity:.4f}")
print(f"Average Specificity: {avg_specificity:.4f}")

```

```python
# ROC Curve and AUC (One-vs-Rest for multi-class)
test_labels_bin = label_binarize(test_labels, classes=range(num_classes))
test_probabilities = svm_classifier.predict_proba(test_features)

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(test_labels_bin[:, i], test_probabilities[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC Curves
plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(num_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve of class {list(test_generator.class_indices.keys())[i]} (AUC = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (One-vs-Rest)')
plt.legend(loc="lower right")
plt.show()
```
