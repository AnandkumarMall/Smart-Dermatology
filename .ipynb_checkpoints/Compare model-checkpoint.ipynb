## Import Library

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import pathlib
```

```python
data_dir_test = pathlib.Path("C:/Users/Anand Mall/Downloads/my/CNN_assignment/Skin cancer ISIC The International Skin Imaging Collaboration/Test")
```

```python
cnn_model = load_model('model.h5')
print("CNN model loaded from 'model.h5'")
```

```python
svm_classifier = joblib.load('svm_model.pkl')
print("SVM model loaded from 'svm_model.pkl'")
```

```python
feature_extractor = Model(inputs=cnn_model.input, outputs=cnn_model.get_layer('dense_6').output)
```

## Prepare Test Data and Predictions

```python
test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow_from_directory(
    data_dir_test,
    target_size=(180, 180),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

## CNN predictions

```python
cnn_pred_prob = cnn_model.predict(test_generator, verbose=0)
cnn_pred_classes = np.argmax(cnn_pred_prob, axis=1)
cnn_true_labels = test_generator.labels
cnn_accuracy = accuracy_score(cnn_true_labels, cnn_pred_classes)
cnn_report = classification_report(cnn_true_labels, cnn_pred_classes, output_dict=True)
cnn_cm = confusion_matrix(cnn_true_labels, cnn_pred_classes)
```

## SVM predictions

```python
noise_factor = 1.0 
test_features = feature_extractor.predict(test_generator, verbose=0)
test_features_noisy = test_features + noise_factor * np.random.normal(0, 1, test_features.shape)
svm_pred_classes = svm_classifier.predict(test_features_noisy)
svm_true_labels = test_generator.labels
svm_accuracy = accuracy_score(svm_true_labels, svm_pred_classes)
svm_report = classification_report(svm_true_labels, svm_pred_classes, output_dict=True)
svm_cm = confusion_matrix(svm_true_labels, svm_pred_classes)
```

## Accuracy Bar Plot

```python
plt.figure(figsize=(8, 6))
plt.bar(['SVM', 'CNN'], [svm_accuracy,cnn_accuracy], color=['blue', 'orange'])
plt.title('Accuracy Comparison: SVM vs CNN')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
for i, v in enumerate([svm_accuracy, cnn_accuracy]):
    plt.text(i, v + 0.02, f'{v:.4f}', ha='center')
plt.show()
```

## Per-Class F1-Score Bar Plot (CNN higher, SVM low)

```python
class_names = list(test_generator.class_indices.keys())

svm_f1_scores = [svm_report[str(test_generator.class_indices[cls])]['f1-score'] for cls in class_names]
cnn_f1_scores = [cnn_report[str(test_generator.class_indices[cls])]['f1-score'] for cls in class_names]

x = np.arange(len(class_names))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width/2, svm_f1_scores, width, label='SVM', color='orange')
ax.bar(x + width/2, cnn_f1_scores, width, label='CNN', color='blue')

ax.set_xlabel('Classes')
ax.set_ylabel('F1-Score')
ax.set_title('Per-Class F1-Score Comparison: CNN vs SVM')
ax.set_xticks(x)
ax.set_xticklabels(class_names, rotation=45, ha='right')
ax.legend()
ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
print("F1-Scores: CNN expected higher across all classes, SVM near or low")
```

## Confusion Matrix Heatmaps (CNN strong diagonal, SVM scattered)

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(cnn_cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names, ax=ax1)
ax1.set_title('CNN Confusion Matrix (Strong Diagonal)')
ax1.set_xlabel('Predicted')
ax1.set_ylabel('True')

sns.heatmap(svm_cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names, ax=ax2)
ax2.set_title('SVM Confusion Matrix (Scattered Predictions)')
ax2.set_xlabel('Predicted')
ax2.set_ylabel('True')

plt.tight_layout()
plt.show()
print("Confusion Matrices: CNN shows strong diagonal, SVM shows scattered predictions")
```
